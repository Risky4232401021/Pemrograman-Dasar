**Operasi yang akan kita lakukan ialah menghitung jumlah detik dari delivery_time menggunakan method total_seconds(). selanjutnya diubah ke dalam satuan hari (dibagi 86400) dan diambil bilangan bulatnya**

```python
delivery_time = orders_df["delivery_date"] - orders_df["order_date"]
delivery_time = delivery_time.apply(lambda x: x.total_seconds())
orders_df["delivery_time"] = round(delivery_time/86400)
```

**Selanjutnya ekspoloring data kita dapat menambahkan kolom baru bernama status pada data customers_df. Kolom ini berisi nilai Active untuk pelanggan yang setidaknya pernah melakukan satu kali order, sedangkan Non Active diberikan kepada pelanggan yang belum pernah melakukan order sama sekali.**

```python
customer_id_in_orders_df =  orders_df.customer_id.tolist()
customers_df["status"] = customers_df["customer_id"].apply(lambda x: "Active" if x in customer_id_in_orders_df else "Non Active")
customers_df.sample(5)
```

**Kode di atas akan menghasilkan lima sample data, Bisa di ubah parameter 5 nya sesuai dengan keinginan untuk melihat data yg lebih banyak**

**Selanjutnya agar kita bisa mendapatkan informasi yang lebih lengkap dari kedua data tersebut, kita perlu menggabungkannya melalui proses join atau merge. Berikut ini merupakan contoh kode yang digunakan untuk melakukan merge pada data orders_df dan customers_df.**

```python
orders_customers_df = pd.merge(
    left=orders_df,
    right=customers_df,
    how="left",
    left_on="customer_id",
    right_on="customer_id"
)
orders_customers_df.head()
```
