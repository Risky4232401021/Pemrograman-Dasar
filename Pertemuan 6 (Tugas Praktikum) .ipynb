```python
# SOAL 2
# Fungsi dengan Parameter Primitif (Matematika)

def hitung_volume_balok(panjang, lebar, tinggi):
    """Menghitung volume balok dari panjang, lebar, dan tingginya."""
    return panjang * lebar * tinggi

# Pemanggilan fungsi
print(f"Volume balok adalah: {hitung_volume_balok(6, 4, 3)}")
```

```python
# SOAL 3
# Fungsi dengan Parameter Non-Primitif (list) dan Perulangan (for)

def hitung_rata_rata(data_nilai):
    """Menghitung rata-rata dengan pengecekan nilai tidak valid."""
    total = 0
    jumlah_valid = 0

    for nilai in data_nilai:
        if 0 <= nilai <= 100:
            total += nilai
            jumlah_valid += 1
        else:
            print(f"Nilai {nilai} tidak valid dan diabaikan.")
    if jumlah_valid > 0:
        return total / jumlah_valid
    return 0

# Pemanggilan:
nilai_praktikan = [85, 75, -10, 90, 110]
print(f"Nilai rata-rata data nilai kelas : {hitung_rata_rata(nilai_praktikan)}")
```

```python
# SOAL 4
# Fungsi dengan Percabangan (if, elif, else)

def tentukan_predikat(nilai):
    """Menentukan predikat (A, B, C, D, E) berdasarkan skor numerik."""

    if not isinstance(nilai, (int, float)):
        return "Error karena nilai tidak berupa angka int dan float."
    if nilai >= 85:
        return "A"
    elif nilai >= 70:
        return "B"
    elif nilai >= 60:
        return "C"
    elif nilai >= 50:
        return "D"
    else:
        return "E"

# Pemanggilan fungsi
print("Nilai 75 =", tentukan_predikat(75))
print("Nilai 48 =", tentukan_predikat(48))
print("Nilai sangat baik =", tentukan_predikat("sangat baik"))
```

```python
# SOAL 5
# Fungsi dengan Implementasi Perulangan while

def hitung_faktorial(n):
    """Menghitung faktorial dari bilangan positif n menggunakan perulangan while."""
    if n < 0:
        return "Error: Faktorial tidak dapat dihitung untuk bilangan negatif."
    elif n == 0:
        return 1
    hasil = 1

    # Perulangan while
    while n > 1:
        hasil *= n
        n -= 1
    return hasil


# Pemanggilan fungsi:
bilangan_n = int(input("Masukkan bilangan bulat positif: "))
print(f"Faktorial dari {bilangan_n} adalah: {hitung_faktorial(bilangan_n)}")
```

```python
# SOAL 6
#Fungsi Komprehensif (Gabungan Perulangan dan Percabangan)
# Analisis Data Suhu Ekstrem

def analisis_suhu(data_suhu, batas_panas=35, batas_dingin=15):
    """Menghitung jumlah hari Panas dan Dingin dari data suhu."""
    panas_count = 0
    dingin_count = 0

    for suhu in data_suhu:
        if suhu >= batas_panas:
            panas_count += 1
        elif suhu <= batas_dingin:
            dingin_count += 1

    return panas_count, dingin_count

# Pemanggilan fungsi:
suhu_mingguan = [36, 30, 14, 25, 38, 16, 28]
jumlah_panas, jumlah_dingin = analisis_suhu(suhu_mingguan)
print(f"Jumlah hari Panas (>=35째C): {jumlah_panas} kali dalam seminggu")
print(f"Jumlah hari Dingin (<=15째C): {jumlah_dingin} kali dalam seminggu")
```
